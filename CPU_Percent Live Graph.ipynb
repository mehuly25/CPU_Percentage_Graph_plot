## CPU Percentage Live Graph 


```python
import matplotlib.pyplot as plt
import psutil
import time
```

```python
psutil.cpu_percent()
```

### Changing the notebook to one that suits live graphs

```python
%matplotlib notebook
plt.rcParams["animation.html"]="jshtml"
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
fig.show()
```

### For Real Time Data

```python
i=0
x,y=[],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y)
    fig.canvas.draw()
    time.sleep(0.1)
    i=i+1
```
